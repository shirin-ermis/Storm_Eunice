# Significance testing

```python
# @jit(nopython=True)
def generate_boot(test, comp, n_boot=10000, sign_lev=0.9, diff_larger_0=True):
    
    '''
    Two sample bootstrap for the mean of two data sets
    '''

    x = test.values  # three dimensional array: number, lat, lon
    y = comp.values

    #Brute force
    x_sign =  np.full((x.shape[1], x.shape[2]), 0)
    for lon in range(x.shape[1]):
        for lat in range(x.shape[2]):
            x_boot = np.random.choice(x[:,lon,lat], size=(n_boot, x.shape[0]), replace=True)
            y_boot = np.random.choice(y[:,lon,lat], size=(n_boot, x.shape[0]), replace=True)

            diff = np.mean(y_boot, axis=1) - np.mean(x_boot, axis=1)
            ci_lower = np.percentile(diff, 1-sign_lev)
            ci_upper = np.percentile(diff, sign_lev)
            if diff_larger_0 & (ci_lower>0):
                x_sign[lon, lat] = 1
            elif (not diff_larger_0) & (ci_upper<0):
                x_sign[lon, lat] = 1
    return x_sign
```

# Maximum wind gusts

```python
# Calculate maximum gusts and minimum pressure fields across extreme composite
maximum_gusts_eps = {}
minimum_pres_eps = {}
mean_pres_eps = {}
for inidate in inidates:
    maximum_gusts_eps[inidate] = {key:value.sel(time = slice('2022-02-18 00', '2022-02-18 23')).fg10.max(dim = ["time"]).mean(dim = "number") for (key,value) in eps[inidate].items()}
    minimum_pres_eps[inidate] = {key:value.sel(time = slice('2022-02-18 00', '2022-02-18 23')).msl.min(dim = ["time"]).mean(dim = "number") for (key,value) in eps[inidate].items()}
    mean_pres_eps[inidate] = {key:value.sel(time = '2022-02-18 12').msl.mean(dim = "number") for (key,value) in eps[inidate].items()}
```
